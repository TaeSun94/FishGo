```
from imageai.Detection.Custom import DetectionModelTrainer
```

```
import os

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
execution_path = os.getcwd() + '/gdrive/My Drive/Colab Notebooks/' # 폴더명에 '_' 들어가있으면 안됨
trainer.setDataDirectory(data_directory=os.path.join(execution_path, "fish"))


trainer.setTrainConfig(object_names_array=["fish"], batch_size=4, num_experiments=200)
trainer.trainModel()
```
